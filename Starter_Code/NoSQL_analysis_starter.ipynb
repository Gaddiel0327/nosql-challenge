### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Define the query to find establishments with London as the Local Authority and RatingValue >= 4
query = {"LocalAuthorityName": "London", "RatingValue": {"$gte": 4}}

# Count the number of documents in the result
count = establishments.count_documents(query)
print(count)

# Find and display the first document in the results using pprint
result = establishments.find_one(query)
pprint(result)
```

```python
# Convert the result to a Pandas DataFrame
result = establishments.find(query)
df = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
print(len(df))

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
# Define the latitude and longitude values
degree_search = 0.01
latitude = 50.769705 
longitude = 0.27694  

# Define the query
query = {
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
    "RatingValue": 5
}

# Define the sort criteria
sort = [("scores.Hygiene", 1)] 

# Define the limit for the number of results
limit = 10

# Perform the search and sort
results = establishments.find(query).sort(sort).limit(limit)

# Print the results
for result in results:
    print(result)
```

```python
# Convert result to Pandas DataFrame
results = establishments.find(query).sort(sort).limit(limit)

# Convert the results to a Pandas DataFrame
df_results = pd.DataFrame(list(results))

# Print the DataFrame
df_results
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Define the pipeline
pipeline = [
    {"$match": {"scores.Hygiene": 0}},  # Match establishments with a hygiene score of 0
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},  # Group matches by Local Authority
    {"$sort": {"count": -1}}  # Sort matches from highest to lowest
]

# Execute the pipeline
results = establishments.aggregate(pipeline)

# Print the number of documents in the result
result_count = sum(1 for _ in results)
print("Number of documents in the result:", result_count)

# Reset the aggregation cursor
results = establishments.aggregate(pipeline)

# Print the first 10 results
print("First 10 results:")
for i, result in enumerate(results):
    if i < 10:
        print(result)

```

```python

# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(establishments.aggregate(pipeline)))

# Display the number of rows in the DataFrame
print(len(df))

# Display the first 10 rows of the DataFrame
df.head(10)
```
